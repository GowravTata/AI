```python
from keras.models import Sequential
from keras.layers import Flatten
from keras.layers import Dense
```

```python
data= pd.read_csv('data.csv')
data.head()
```

```python
x_input=data[['x1','x2']]
y_label=data[['label']]
```

```python
x_input
```

```python
y_label
```

```python
plt.scatter(data.x1,data.x2,c=data.label)
```

```python
mymodel=Sequential()
```

```python
mymodel.add(Dense(units=1,input_dim=2,activation='sigmoid'))
```

```python
mymodel.summary()
```

```python
mymodel.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
mymodel.fit(x_input,y_label,epochs=1000)
```

```python
mymodel.evaluate(x_input,y_label)
```
